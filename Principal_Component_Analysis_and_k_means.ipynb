```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Step 2: Create fake data (like customers)
np.random.seed(0)

group1 = np.random.normal(loc=[5, 5, 5], scale=1, size=(50, 3))
group2 = np.random.normal(loc=[10, 10, 10], scale=1, size=(50, 3))
group3 = np.random.normal(loc=[15, 5, 10], scale=1, size=(50, 3))

data = np.vstack((group1, group2, group3))

# Step 3: Apply K-means (grouping)
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(data)

# Step 4: Apply PCA (reduce to 2D)
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data)

# Step 5: Plot result
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels)
plt.title("PCA + K-means Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
